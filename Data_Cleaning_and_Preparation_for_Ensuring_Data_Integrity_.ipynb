```python
df = pd.read_excel("Dataset for Data Analytics.xlsx")

df.head()
```

# Step 5: Find Missing Values

```python
df.isnull().sum()
```

```python
missing=(df.isnull().sum()/len(df))*100

print(missing)
```

# Step 6: Handle Missing Values

```python
num_cols=df.select_dtypes(include=np.number)

for col in num_cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
cat_cols=df.select_dtypes(include='object')

for col in cat_cols:
    df[col]=df[col].fillna(df[col].mode()[0])
```

# Step 7: Find Duplicate Records

```python
duplicates=df.duplicated().sum()

print("Duplicates:",duplicates)
```

# Step 8: Remove Duplicates

```python
df=df.drop_duplicates()

print(df.shape)
```

# Step 9: Check Duplicate IDs

```python
duplicate_ids = df["OrderID"].duplicated().sum()

print(duplicate_ids)
```

# Step 7: Check Data Types

```python
print(df.dtypes)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

print(df["Date"].head())
```

# Step 8: Remove Extra Spaces (Text Cleaning)

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = df[col].str.strip()

print("Spaces Removed")
```

# Step 9: Standardize Text Format

```python
df["Product"] = df["Product"].str.title()

df["OrderStatus"] = df["OrderStatus"].str.title()
```

```python

print(df[["Product","OrderStatus"]].head())
```

# Step 10: Verify No Duplicate Order IDs

```python
duplicate_ids = df["OrderID"].duplicated().sum()

print("Duplicate Order IDs:", duplicate_ids)
```

# Step 11: Verify Date Format

```python
print(df["Date"].dtype)
```

# Step 12: Save Cleaned Dataset

```python
df.to_excel("Cleaned_Dataset.xlsx",index=False)

print("Dataset Saved Successfully")
```
